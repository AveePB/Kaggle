<h2>Titanic - Model Boosting</h2>

<h3>Importing libraries</h3>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<h3>Loading training & testing datasets</h3>

```python
df_train = pd.read_csv("./data/train.csv")
```

```python
df_train
```

```python
X_train = df_train.iloc[:, df_train.columns != "Survived"].values
y_train = df_train.loc[:, "Survived"].values
```

```python
X_train
```

```python
y_train
```

<h3>K-Neighbors Classifier</h3>

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
grid_search_knn = GridSearchCV(knn, param_grid_knn, cv=5, scoring='accuracy', verbose=2)
grid_search_knn.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search_knn.best_params_)
print("Best Score:", grid_search_knn.best_score_)
```

<h3>Decision Tree Classifier</h3>

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()

param_grid_tree = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 10, 20, 30, 40, 50],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 3, 4]
}

grid_search_tree = GridSearchCV(tree, param_grid_tree, cv=5, scoring="accuracy", verbose=2)
grid_search_tree.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search_tree.best_params_)
print("Best Score:", grid_search_tree.best_score_)
```

<h3>Support Vector Classifier</h3>

```python
from sklearn.svm import SVC

svc = SVC()

param_grid_svc = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid']
}

grid_search_svc = GridSearchCV(svc, param_grid_svc, cv=5, scoring="accuracy", verbose=2)
grid_search_svc.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search_svc.best_params_)
print("Best Score:", grid_search_svc.best_score_)
```

<h3>Conclusion</h3>
